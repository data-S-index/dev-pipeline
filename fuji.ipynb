```python
import base64
import requests
from datetime import datetime
import json

# FUJI API configuration
PORT = 54001
FUJI_ENDPOINT = f"http://localhost:{PORT}/fuji/api/v1/evaluate"

USERNAME = "marvel"
PASSWORD = "wonderwoman"

BASIC_AUTH = base64.b64encode(f"{USERNAME}:{PASSWORD}".encode("ascii")).decode("ascii")
HEADERS = {
    "accept": "application/json",
    "Authorization": f"Basic {BASIC_AUTH}",
    "Content-Type": "application/json",
}

# Test with one DOI
TEST_DOI = "10.5061/dryad.mm46r/7"  # Replace with a real DOI for testing

print(f"Testing with DOI: {TEST_DOI}")
print(f"Endpoint: {FUJI_ENDPOINT}")
```

```python
# Make the request
payload = {
    "object_identifier": TEST_DOI,
}

print("Making request to FUJI API...")
response = requests.post(FUJI_ENDPOINT, json=payload, headers=HEADERS, timeout=60)
response.raise_for_status()
results = response.json()

print("âœ… Request successful!")
print(f"Response status: {response.status_code}")
```

```python
summary = results.get("summary", {})
score_percent = summary.get("score_percent", {})
score = score_percent.get("FAIR")
evaluation_date = results.get("end_timestamp")

print(f"FAIR Score: {score}")
print(f"Full response structure:")
print(f"Evaluation date: {evaluation_date}")
print(f"  - Has summary: {'summary' in results if results else False}")
print(f"  - Score: {score}")
```
